```python
#financial cost of USA gov departments for the year 2023
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Usa_budget_2023.csv")
data.head()
```

```python
print(data)
```

```python
#print just one column of the agency so that only the important departments are shown
print(data[['Agency Name']].to_string(index=False))
```

```python
#now let us just print the agency and the corrensponding financial data
#what I did was call each agency by its index position

data = data.iloc[[0,1,3,5,6,8,10,11,12,13,14,17,19,21,24],:]
print(data)
```

```python
#now we plot the indexed data from above to look at the spending of the us government for the year 2023
data.plot.bar(x='Agency Name',y='Gross Cost (in Billions)')
```

```python
#We can see that the department of health and human services cost the us government the most in 2023. 
#so we will try a donut plot to have a deeper look of the costs amoung the departments.

df = data["Gross Cost (in Billions)"]
labels = data["Agency Name"]
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of Spending", fontsize=20)
plt.show()
```
